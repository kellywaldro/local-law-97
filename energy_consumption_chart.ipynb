## Plotting emissions by sector

Using NYC data downloaded from the Mayor's Office of Climate & Environmental Justice 

Light cleaning and plotting using Altair:

```python
import pandas as pd
```

```python
df = pd.read_csv("./raw data/consumption_2021.csv")
```

```python
df
```

```python
df['FY 2021 tCO2e'] = df['FY 2021 tCO2e'].str.replace(",", "")
```

```python
df['FY 2021 tCO2e'] = df['FY 2021 tCO2e'].astype(float)
```

```python
grouped = df.groupby('Main Sector')['FY 2021 tCO2e'].sum().reset_index()
```

```python
grouped
```

```python
grouped = grouped[grouped['Main Sector'] != 'Total']
```

```python
grouped
```

```python
grouped = grouped.sort_values(by='FY 2021 tCO2e', ascending=False)
```

```python
grouped["percent"] = grouped["FY 2021 tCO2e"] * 100 / grouped["FY 2021 tCO2e"].sum()
```

```python
grouped
```

This shows me the percentage of energy consumption that each sector is responsible for. Needed to verify the "buildings account for 70 percent of the city's emissions" statement. 

```python
import altair as alt

# make a bar chart of FY 2021 tCO2e

alt.Chart(grouped).mark_bar().encode(
    y=alt.Y('Main Sector', sort='-x'),
   x='FY 2021 tCO2e',
    
)
```
